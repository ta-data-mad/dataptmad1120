#### 1. Use a list comprehension to create and print a list of consecutive integers starting with 1 and ending with 50.

#### 2. Use a list comprehension to create and print a list of even numbers starting with 2 and ending with 200.

```python
even_numbers = [i for i in range(2, 201) if i%2==0]
even_numbers[0:20]
#wrote an index from element 0 to 20 on the list to see the result clearly.
```

#### 7. Use a list comprehension to select and print the names of all CSV files in the */data* directory.

```python
#just found it in stack overflow because i didn't know how to do this one and don't know how to pack this in a list comprehension. 

from os import listdir

def find_csv_filenames(path_to_dir, suffix=".csv"):
    filenames = listdir(path_to_dir)
    return [filename for filename in filenames if filename.endswith(suffix)]

filenames = find_csv_filenames("/home/emily/Bootcamp/dataptmad1120/module-1/list-comprehension/data")
for name in filenames:
    print(name)
```
