#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userID, ViewCount, CommentCount

```python
users_columns = pd.read_sql_query("SELECT Id as userId,Reputation,Views,UpVotes, DownVotes FROM users",engine)
#users_columns
df_users_columns = pd.DataFrame(users_columns)
df_users_columns

posts_columns = pd.read_sql_query("SELECT Id as postId,Score,OwnerUserId as userId,ViewCount, CommentCount FROM posts",engine)
#posts_columns
df_posts_columns = pd.DataFrame(posts_columns)
#df_posts_columns
df_users_columns
                    
```

#### 9. Merge the new dataframes you have created, of users and posts. 
You will need to make an inner [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
inner = pd.merge(df_users_columns, df_posts_columns, how = "inner")
inner
```
