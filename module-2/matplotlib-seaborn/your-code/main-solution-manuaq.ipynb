#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x = np.linspace(0,1,100)
y_nor = x**2
y_log = np.exp(x)

fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(16, 8))

axes[0].plot(x, y_nor, '.-.', label='linear', color='green')
axes[0].set_title('Plotting y = x^2 - Normal Scale')
axes[0].set_ylabel('TBD')
axes[0].set_xscale('linear')
axes[0].set_yscale('linear')


axes[1].plot(x, y_log, '-', label='Logarithmic')
axes[1].set_title('Plotting y = e^x - Logarithmic scale')
axes[1].set_ylabel('TBD2')
axes[1].set_xscale(value = 'log')
axes[1].set_yscale(value = 'log')


plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(12,4))


ax.set_title('Total Fitbit Steps by Month')
ax.bar(monthly_steps["Months"], monthly_steps["Steps"],color= "DarkTurquoise", label='Total Steps')
ax.text(9, 305000, 'Median Steps by Month')
ax.axhline(monthly_steps["Steps"].median(), color="Red")
ax.xaxis.set_tick_params(rotation=20)
ax.legend()


# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(12,4))
ax.set_title('Aggregated Number of steps Workdays vs. Weekends')
ax.bar(work_or_weekend["Work_or_Weekend"], work_or_weekend["Steps"],color= "DarkOrange", label='Total Steps')
ax.text(0, 2000000, 'Median Steps')
ax.axhline(work_or_weekend["Steps"].median(), color="Red")
ax.xaxis.set_tick_params(rotation=20)
ax.legend()


fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(12,4))
ax.set_title('Average steps on a Workday vs. Weekend day')
ax.bar(work_or_weekend["Work_or_Weekend"], [work_or_weekend.loc[0,'Steps']/2, work_or_weekend.loc[1,'Steps']/5],color= "Navy", label='Total Steps')
ax.axhline(0, 500000, color="Red")
ax.xaxis.set_tick_params(rotation=20)
ax.grid(True)
ax.set_ylim([520000,535000])
ax.legend()

```

#### Write a loop to plot 3 scatter plots of the following features:

#### Plot the correlation matrix using seaborn.

```python
# your code here

titanic.corr()
#plt.figure(figsize=(16, 8))
heatmap = sns.heatmap(titanic.corr(), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

```python
np.triu(np.ones_like(titanic.corr()))

plt.figure(figsize=(16, 6))
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(titanic.corr(), dtype=np.bool))
heatmap = sns.heatmap(titanic.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='seismic')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```
